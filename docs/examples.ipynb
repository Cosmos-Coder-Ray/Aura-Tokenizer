# Aura-Tokenizer Examples

This notebook demonstrates the usage of the `aura-tokenizer` library.

```python
from aura_tokenizer import Tokenizer

# Initialize
tokenizer = Tokenizer()
```

```python
# Train (assuming corpus.txt exists)
# with open("corpus.txt", "w") as f: f.write("Hello world! This is a test corpus.")
# tokenizer.train(["corpus.txt"], 1000)
```

```python
# Configure
tokenizer.set_normalization_form("NFC")
tokenizer.set_lowercase(True)
tokenizer.create_bert_post_processor(True)
```

```python
# Encode
encoding = tokenizer.encode("Hello World")
print(encoding.input_ids)
```
